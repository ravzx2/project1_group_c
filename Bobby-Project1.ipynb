```python
import csv
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
from datetime import date
import random
```

```python
years = [2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017]
datelist = []
```

```python
def rdate():
    for y in years:
        start_dt = date(year=y, day=1, month=1).toordinal()
        end_dt = date(year=y, day=31, month=12).toordinal()
        random_day = date.fromordinal(random.randint(start_dt, end_dt))
        #print(random_day)

        if random_day not in datelist:
            datelist.append(random_day)

    #print(datelist)
```

```python
for x in range(100):
    rdate()
```

```python
print(f"datelist = {len(datelist)}")
```

```python
event_df = pd.DataFrame({'Date':datelist})
event_df['Product'] = ""
event_df['Outcome'] = ""
event_df['Gender'] = ""
event_df['Age'] = ""
event_df.head(10)
```

```python
url = f"https://api.fda.gov/food/event.json"
term = f"date_started"
date = f"2008-07-11"
query_url = f"{url}?search={term}:{date}&limit=1"

event_response = requests.get(query_url)
event_json = event_response.json()
print(event_json)
```

```python
count = 1
print(f"Beginning Data Retrieval")
print(f"-----------------------------")
for index, row in event_df.iterrows():
    try:
        url = f"https://api.fda.gov/food/event.json"
        term = f"date_started"
        date = event_df.loc[index, 'Date']
        
        query_url = f"{url}?search={term}:{date}&limit=1"

        event_response = requests.get(query_url)
        event_json = event_response.json()
        print(f"Processing record {count} for {event_df.loc[index, 'Date']}")
        count += 1
        
        event_df.loc[index, 'Product'] = event_json['products']['industry_name']
        event_df.loc[index, 'Outcome'] = event_json['outcomes']
        event_df.loc[index, 'Gender'] = event_json['consumer']['gender']
        event_df.loc[index, 'Age'] = event_json['consumer']['age']
    
    except KeyError:
        print(f"Missing information. Skipping...")

print(f"-----------------------------")
print(f"Data Retrieval Complete")
print(f"-----------------------------")
```
