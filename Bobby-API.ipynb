```python
import csv
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import datetime as dt
import random
```

```python
years = [2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017]
datelist = []
```

```python
def rdate():
    for y in years:
        start_dt = dt.date(year=y, day=1, month=1).toordinal()
        end_dt = dt.date(year=y, day=31, month=12).toordinal()
        random_day = dt.date.fromordinal(random.randint(start_dt, end_dt))
        #print(random_day)

        if random_day not in datelist:
            datelist.append(random_day)

    #print(datelist)
```

```python
for x in range(200):
    rdate()
```

```python
print(f"datelist = {len(datelist)}")
```

```python
print(datelist)
```

```python
event_df = pd.DataFrame({'Date':[], 'Product':[], 'Reactions':[], 'Number of Reactions':[], 'Outcome':[], 'Number of Outcomes':[], 'Gender':[], 'Age':[]})
event_df.head(10)
```

```python
event_df['Outcome'] = event_df['Outcome'].astype(object)
event_df['Reactions'] = event_df['Reactions'].astype(object)
```

```python
url = f"https://api.fda.gov/food/event.json"
term = f"date_started"
date = f"2008-08-11"
query_url = f"{url}?search={term}:{date}&limit=10"

event_response = requests.get(query_url)
event_json = event_response.json()
print(event_json)
```

```python
count = 1
rowcount = 0
index = datelist[0]
print(f"Beginning Data Retrieval")
print(f"-----------------------------")
for index in range(len(datelist)):
    try:
        url = f"https://api.fda.gov/food/event.json"
        term = f"date_started"
        date = datelist[index].strftime('%Y%m%d')
        
        query_url = f"{url}?search={term}:{date}&limit=10"
        #print(query_url)
        event_response = requests.get(query_url)
        event_json = event_response.json()
        print(f"Processing record {count} for {datelist[index]}")
        count += 1
        #print(event_json['results'][0]['consumer']['age'])
        for i in range(0, len(event_json['results'])):   
            event_df.loc[rowcount, 'Date'] = event_json['results'][i]['date_started']
            event_df.loc[rowcount, 'Product'] = event_json['results'][i]['products'][0]['industry_name']
            event_df.at[rowcount, 'Outcome'] = event_json['results'][i]['outcomes']
            event_df.at[rowcount, 'Reactions'] = event_json['results'][i]['reactions']
            event_df.loc[rowcount, 'Gender'] = event_json['results'][i]['consumer']['gender']
            event_df.loc[rowcount, 'Age'] = event_json['results'][i]['consumer']['age']
            rowcount += 1
    except KeyError:
        print(f"Missing information. Skipping...")

print(f"-----------------------------")
print(f"Data Retrieval Complete")
print(f"-----------------------------")
```
